```python
df = pd.read_csv('/content/drive/MyDrive/lenta-ru-news.csv')
```

```python
selected_topics = df[~df['topic'].isin(['Россия','Мир','69-я параллель','Крым','Культпросвет','Легпром', 'Библиотека','Оружие','ЧМ-2014','МедНовости', 'Сочи','Культпросвет '])]
```

```python
selected_topics['topic'].value_counts ()
```

```python
selected_topics.shape
```

```python
clearedNA = selected_topics.dropna()
```

```python
clearedNA.shape
```

```python
text_topic_df = clearedNA[['text','topic']]
```

```python
df_1000_items = text_topic_df.sample (n= 10000)
```
